```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
import math
import random

def calculate_area_of_circle(radius):
    return math.pi * radius ** 2

def calculate_total_area_of_region():
    return 40 * 40

def calculate_total_area_covered(radii):
    total_area_covered = sum(calculate_area_of_circle(radius) for radius in radii)
    return total_area_covered

def calculate_total_uncovered_area(total_area_region, total_area_covered):
    return total_area_region - total_area_covered

def display_results(coordinates, radii):
    # Display coordinates of sensors
    print("Coordinates of sensors:")
    for i, coord in enumerate(coordinates, 1):
        print(f"Sensor {i}: {coord}")

    # Calculate areas
    total_area_region = calculate_total_area_of_region()
    total_area_covered = calculate_total_area_covered(radii)
    total_uncovered_area = calculate_total_uncovered_area(total_area_region, total_area_covered)

    # Display areas
    print("Total area of the region:", total_area_region)
    print("Total area covered by circles:", total_area_covered)
    print("Total area uncovered by circles:", total_uncovered_area)

def plot_circles(coordinates, radii):
    # Create figure and axis
    fig, ax = plt.subplots()

    # Set axis limits
    ax.set_xlim(0, 40)
    ax.set_ylim(0, 40)

    # Plot circles
    for i, (coord, radius) in enumerate(zip(coordinates, radii), 1):
        circle = Circle(coord, radius, alpha=0.5, label=f'Circle {i}')
        ax.add_patch(circle)

    # Plot points
    plt.scatter(*zip(*coordinates), color='red', label='Centers')

    # Add labels and title
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Best Positions given by PSO Algorithm')

    # Show plot
    plt.grid(True)
    plt.gca().set_aspect('equal', adjustable='box')
#     plt.legend()
    plt.show()

# Given coordinates
coordinates = [(6,5.6),(21.5,9.5),(35,7),(21,30),(6,33),(35,34),(8,19),(34,21)]

# Given radii
radii = [6,10,6,9.5,6.5,6,9,6.7]

# Plot circles
plot_circles(coordinates, radii)

# Display results
display_results(coordinates, radii)
```
